**Filtering Data frames**

```python
#filtering is about rows
stats.head()
```

```python
#in which row in the specified column is the condition met?
stats.InternetUsers < 2 
```

```python
Filter = stats.InternetUsers < 2
Filter
```

```python
#recall slicing
stats[30:40]
```

```python
#note, you can also specific the true/false value
# here we created 'Filter where InternetUsers <2'
stats[Filter] #conceptually like R
```

```python
#let's use the above in practice
Filter2 = stats.BirthRate > 40
Filter2
```

```python
Filter2 = stats.BirthRate > 40
stats[Filter2]
```

```python
#an even more interesting concepts:
stats[stats.BirthRate > 40 and stats.InternetUsers < 2] #error why?
```

```python
#where is the error?
stats[stats.BirthRate > 40] #python is not a vectorize programming language
```

```python
# the 'and' doesn't know how to work with the objects
# here python wants a single value for each of the object (series) above
#but this is not what we want. What do we do?

Filter & Filter2
```

```python
stats[Filter & Filter2]
```

```python
stats[stats.BirthRate > 40 & stats.InternetUsers < 2] #why?
#because we need brackets to speciy the order of operations
```

```python
stats[(stats.BirthRate > 40) & (stats.InternetUsers < 2)] #conceptually similar to R
```

**Visualizing with Seaborn: Part1**

```python
#press 'shift +tab' to get  more information about the function
 # doesn't look like linear relationship -> remove regression line
vis3 = sns.lmplot(data=stats, x ='InternetUsers', y='BirthRate', fit_reg=False)
```

**Keyword Arguements in Python (Advance)**

```python
#this is a conceptual principle in python

#Marker size: how do we change the colour  and size of marker?
#keyword arguements are arguements that can be passed into functions via dictionary
#kws: keywords
vis5 = sns.lmplot(data=stats, x ='InternetUsers', y='BirthRate', fit_reg=False,
                 hue='IncomeGroup', size =7, scatter_kws={"s":200} ) #this is a dictionary with 1 element
```
